# Wind Dataset Cleaning Workbook

This is a separate tab/workbook for the ZIP dataset cleanup.

Generated files in this folder:
- `wind_spatiotemporal_cleaned.csv`
- `wind_spatiotemporal_cleaned_imputed.csv`
- `wind_turbine_coordinates.csv`


```python
import pandas as pd
from pathlib import Path

base = Path(".")
clean = pd.read_csv(base / "wind_spatiotemporal_cleaned.csv")
imputed = pd.read_csv(base / "wind_spatiotemporal_cleaned_imputed.csv")
coords = pd.read_csv(base / "wind_turbine_coordinates.csv")

print("clean shape:", clean.shape)
print("imputed shape:", imputed.shape)
print("coords shape:", coords.shape)
print("time range:", clean["Time"].iloc[0], "->", clean["Time"].iloc[-1])
```

```python
missing_clean = int(clean.iloc[:, 1:].isna().sum().sum())
missing_imputed = int(imputed.iloc[:, 1:].isna().sum().sum())

summary = pd.DataFrame(
    {
        "dataset": ["clean", "imputed"],
        "rows": [len(clean), len(imputed)],
        "cols": [clean.shape[1], imputed.shape[1]],
        "missing_values": [missing_clean, missing_imputed],
    }
)
summary
```
